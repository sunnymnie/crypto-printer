```python
from IPython.display import Image, display
def backtest(a, b):
    return Image(filename=f'../data/generated/backtest-graphs/{a.upper()}-{b.upper()}.png') 
def forwardtest(a, b):
    return Image(filename=f'../data/generated/forwardtest-graphs/{a.upper()}-{b.upper()}.png') 
def show(a, b):
    listOfImageNames = [f'../data/generated/backtest-graphs/{a.upper()}-{b.upper()}.png',
                        f'../data/generated/forwardtest-graphs/{a.upper()}-{b.upper()}.png']
    for imageName in listOfImageNames:
        display(Image(filename=imageName, width = 500, height = 200))
```
